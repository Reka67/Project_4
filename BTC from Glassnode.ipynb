```python
new_addresses = gn.get('https://api.glassnode.com/v1/metrics/addresses/new_non_zero_count', a="BTC", s='2020-11-01' ,u='2020-11-21')
```

```python
total_addresses = gn.get('https://api.glassnode.com/v1/metrics/addresses/count', a="BTC", s='2020-11-01' ,u='2020-11-21')
```

```python
start_date = '2018-01-02'
end_date = '2020-11-20'

price = gn.get('https://api.glassnode.com/v1/metrics/market/price_usd_close',a="BTC", s=start_date ,u=end_date)
active_addresses = gn.get('https://api.glassnode.com/v1/metrics/addresses/active_count',a="BTC", s=start_date ,u=end_date)
new_addresses = gn.get('https://api.glassnode.com/v1/metrics/addresses/new_non_zero_count', a="BTC", s=start_date ,u=end_date)
total_addresses = gn.get('https://api.glassnode.com/v1/metrics/addresses/count', a="BTC", s=start_date ,u=end_date)


new_adddresses_list = [i for i in new_addresses]
active_addresses_list = [i for i in active_addresses]
total_addresses_list = [i for i in total_addresses]
price_list = [i for i in price]
```

```python
data = pd.DataFrame()
```

```python
date = price.index
data["date"] = date
data["total_addresses"] = total_addresses_list
data["active_addresses"] = active_addresses_list
data["new_addresses"] = new_adddresses_list
data["price"] = price_list
```
