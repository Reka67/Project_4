```python
def changes (df):
    change = []
    for i in range(len(df)):
        if i == 0:
            x = 0
        else:
            x = (df["Close"][i] - df["Close"][i-1]) / df["Close"][i-1]

        change.append(x)

    df["Change"] = change
```
